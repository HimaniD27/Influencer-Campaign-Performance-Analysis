```python
engagement_metrics= {'Influencer ID': posts['Influencer ID'], 'Platform': posts['Platform'], 'Reach': posts['Reach'], 'Likes': posts['Likes'], 'Comments': posts['Comments']}
engagement_df= pd.DataFrame(engagement_metrics)
engagement_df
```

# Merging all the datasets

```python
tracking_data.head(2)
```

```python
payouts.head(2)
```

```python
posts.head(2)
```

```python
influencers.head(2)
```

```python
df1= payouts.merge(tracking_data, on=['Influencer ID','Orders'], how='inner')
df1.head()
```

```python
df2= df1.merge(posts, on=['Influencer ID', 'Date'], how='inner')
df2.head()
```

```python
influencers= influencers.rename(columns={'ID':'Influencer ID'})
influencers.head()
```

```python
df= df2.merge(influencers, on=['Influencer ID','Platform'], how='inner')
```

```python
df.head()
```

#

# ROI & ROAS Metrics 

### Return On Ad Spend (ROAS)

```python
df_main= df.groupby(by='Influencer ID')[['Revenue', 'Total Payout']].sum()

df_main.head()
```

```python
df_main['Return On Ad Spend (ROAS)']= df_main['Revenue']/df_main['Total Payout']
```

#### Influencers with high ROAS

```python
df_main.sort_values(by='Return On Ad Spend (ROAS)', ascending=False).head(10)
```

#### Influencers with low ROAS

```python
df_main.sort_values(by='Return On Ad Spend (ROAS)', ascending=False).tail(10)
```

--- 

### Incremental ROAS 

```python
# Let us assume that the baseline revenue is 2000

baseline_revenue= 2000
```

```python
df_main['Incremental Revenue']= df_main['Revenue']- baseline_revenue
```

```python
df_main['Incremental ROAS']= df_main['Incremental Revenue']/df_main['Total Payout']
```

#### Influencers with high Incremental ROAS

```python
df_main.sort_values(by='Return On Ad Spend (ROAS)', ascending=False).head(10)
```

#### Influencers with low ROAS

```python
df_main.sort_values(by='Return On Ad Spend (ROAS)', ascending=False).tail(10)
```

--- 

### Net Revenue

```python
df_main['Net Revenue']= df_main['Revenue']-df_main['Total Payout']
```

#### Influencers with high Net Revenue

```python
df_main.sort_values(by='Net Revenue', ascending=False).head(10)
```

#### Influencers with low Net Revenue

```python
df_main.sort_values(by='Net Revenue', ascending=False).tail(10)
```

--- 

### Return On Investment (%)

```python
df_main['Return On Investment (ROI %)']= df_main['Net Revenue']/df_main['Total Payout']*100
```

#### Influencers with high ROI

```python
df_main.sort_values(by='Return On Investment (ROI %)', ascending=False).head(10)
```

#### Influencers with low ROI

```python
df_main.sort_values(by='Return On Investment (ROI %)', ascending=False).tail(10)
```

#

```python
merged_df= df.merge(df_main, on=['Influencer ID','Revenue','Total Payout'], how='inner')
```

```python
merged_df['Engagement rate (%)']=((merged_df['Likes']+merged_df['Comments'])/engagement_df['Reach'])*100
merged_df['Like rate (%)']=merged_df['Likes']/merged_df['Reach']*100
merged_df['Comment rate (%)']=merged_df['Comments']/merged_df['Reach']*100
```

#

# Influencer Quality & Efficiency Metrics

### Revenue per Follower

```python
merged_df['Revenue per Follower']=merged_df['Revenue']/merged_df['Follower Count']
merged_df.head(2)
```

### Orders per 10k Followers

```python
merged_df['Orders per 10k Followers'] = merged_df['Orders'] / (merged_df['Follower Count'] / 10000)
merged_df.head(2)
```

### Engagement ROI

```python
merged_df['Engagement ROI'] = merged_df['Return On Ad Spend (ROAS)'] * merged_df['Engagement rate (%)']
merged_df.head(2)
```
